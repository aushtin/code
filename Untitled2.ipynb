```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import matplotlib.cm as cmap
import random
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.naive_bayes import  GaussianNB, MultinomialNB
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.linear_model import  LogisticRegression
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.feature_selection import RFE
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.multiclass import OneVsRestClassifier
import time
import gc
import logging
from sklearn.metrics import accuracy_score
#import lightgbm as lgb

#split data
def split_dataset(split_size, ts_df, metadata_df):
    
    print('Splitting dataset of size', metadata_df.shape[0])
    labels = pd.DataFrame(metadata_df.iloc[:, 11], columns = ['target'])
    metadata_df = metadata_df.drop(['target'], axis = 1)

    metadata1, metadata2, labels1, labels2 = train_test_split(metadata_df, labels, test_size = split_size, random_state = 888) 

    metadata1.loc[:, 'target'] = labels1['target']
    metadata2.loc[:, 'target'] = labels2['target']

    time_series1 = ts_df.loc[ts_df['object_id'].isin(metadata1.loc[:, 'object_id'])]
    time_series2 = ts_df.loc[ts_df['object_id'].isin(metadata2.loc[:, 'object_id'])]

    print('The size of the first split is:', metadata1.shape[0])
    print('Check number of object ids in time series for first split:', time_series1['object_id'].value_counts().shape[0])
    print('The size of the second split is:', metadata2.shape[0])
    print('Check number of object ids in time series for second split:', time_series2['object_id'].value_counts().shape[0])

    return metadata1, time_series1, metadata2, time_series2

def plot_class_dist(metadata):
    
    #visualize data in pretraining set

    #metadata
    class_counts = metadata['target'].value_counts()

    #we know that the metadata provides a feature that identifies objects as intergalactic (inside our galaxy) and extragalactic (outside our galaxy).  objects. 

    ig = metadata.loc[metadata['hostgal_specz'] == 0]
    ig_class_counts = ig['target'].value_counts()

    #display(ig_class_counts)

    eg = metadata.loc[metadata['hostgal_specz'] != 0]
    eg_class_counts = eg['target'].value_counts()

    #display(eg_class_counts)

    total_classes = ig_class_counts.size + eg_class_counts.size

    plt.bar(range(0, ig_class_counts.size), ig['target'].value_counts())
    plt.bar(range(ig_class_counts.size, total_classes), eg['target'].value_counts())
    plt.xticks(range(0, total_classes), ig_class_counts.index.tolist() + eg_class_counts.index.tolist())
    # plt.xticks(range(ig_class_counts.size, total_classes), eg_class_counts.index)
    plt.xlabel('Class')
    plt.ylabel('# of Occurrences')
    plt.title('Objects Class Occurrence Count')
    plt.legend(['Intergalactic', 'Extragalactic'])
    
    plt.savefig('./figures/classcount.png')

def plot_random_light_curves(time_series):
    
    g = time_series.groupby('object_id')
    pt_obj_list = []
    for name, group in g:
        pt_obj_list.append(name)

    plt.rcParams['figure.figsize'] = [15, 8]

    obj_id1 = random.choice(pt_obj_list)
    obj_id = obj_id1
    obj = g.get_group(obj_id)
    obj1 = obj.copy()
    fig, ax = plt.subplots()
    ax = obj.plot.scatter(x = 'mjd', y = 'flux', c = 'passband', cmap = cmap.get_cmap('Set1'), title = 'Object ID ' + str(obj_id), ax = ax )
    ax.set_xlabel('mjd')
    ax.set_ylabel('flux')
    plt.savefig('./figures/lc1.png')
    
    obj_id2 = random.choice(pt_obj_list)
    obj_id = obj_id2
    obj = g.get_group(obj_id)
    obj2 = obj.copy()
    fig, ax = plt.subplots()
    ax = obj.plot.scatter(x = 'mjd', y = 'flux', c = 'passband', cmap = cmap.get_cmap('Set1'), title = 'Object ID ' + str(obj_id), ax = ax )
    ax.set_xlabel('mjd')
    ax.set_ylabel('flux')
    plt.savefig('./figures/lc2.png')
    
    obj_id3 = random.choice(pt_obj_list)
    obj_id = obj_id3
    obj = g.get_group(obj_id)
    obj3 = obj.copy()
    fig, ax = plt.subplots()
    ax = obj.plot.scatter(x = 'mjd', y = 'flux', c = 'passband', cmap = cmap.get_cmap('Set1'), title = 'Object ID ' + str(obj_id), ax = ax )
    ax.set_xlabel('mjd')
    ax.set_ylabel('flux')
    plt.savefig('./figures/lc3.png')

    obj_id4 = random.choice(pt_obj_list)
    obj_id = obj_id4
    obj = g.get_group(obj_id)
    obj4 = obj.copy()
    fig, ax = plt.subplots()
    ax = obj.plot.scatter(x = 'mjd', y = 'flux', c = 'passband', cmap = cmap.get_cmap('Set1'), title = 'Object ID ' + str(obj_id), ax = ax )
    ax.set_xlabel('mjd')
    ax.set_ylabel('flux')
    plt.savefig('./figures/lc4.png')

    obj_id5 = random.choice(pt_obj_list)
    obj_id = obj_id5
    obj = g.get_group(obj_id)
    obj5 = obj.copy()
    fig, ax = plt.subplots()
    ax = obj.plot.scatter(x = 'mjd', y = 'flux', c = 'passband', cmap = cmap.get_cmap('Set1'), title = 'Object ID ' + str(obj_id), ax = ax )
    ax.set_xlabel('mjd')
    ax.set_ylabel('flux')
    plt.savefig('./figures/lc6.png')

    #return object ids and objs in case we want to use later
    return [obj_id1, obj_id2, obj_id3, obj_id4, obj_id5], [obj1, obj2, obj3, obj4, obj5]

#unused. originally I was normalizing all of the time series, but this did not seem to make a big difference
def normalize_ts(ts, metadata):

    normalized_ts = ts.copy()
    obj_ids = metadata.loc[:, 'object_id']
    n_objects = len(obj_ids)
    
    print('Starting normalization of', n_objects, 'object time series...', end = '')

    for i, id in enumerate(obj_ids):
        for pb in range(0, 6):
            series = normalized_ts.loc[ (normalized_ts.loc[:, 'object_id'] == id), 'flux']
            mean = series.mean()
            std = series.std()
            normalized_ts.loc[ (normalized_ts.loc[:, 'object_id'] == id), 'flux'] = (series - mean) / std
        #print(i+1, 'out of', n_objects, 'done...')
    
    print('done.', n_objects, 'objects have had their time series normalized.')
    
    return normalized_ts

def extract_pbts_features (obj_id, df, feats = ['mean', 'stddev', 'med', 'max', 'min', 'kurt', 'skew', 'diff']):

    feature_list = [obj_id]
    for i, f in enumerate(feats):
        for name, obj_ts in df.groupby('passband'):
            ts = obj_ts.loc[:, 'flux'].copy()
            if f == 'mean':
                feature_list.append(ts.mean())
            elif f == 'stddev':
                feature_list.append(ts.std())
            elif f == 'med':
                feature_list.append(ts.median())
            elif f == 'max':
                feature_list.append(ts.max())
            elif f == 'min':
                feature_list.append(ts.min())
            elif f == 'kurt':
                feature_list.append(ts.kurtosis())
            elif f == 'skew':
                feature_list.append(ts.skew())
            elif f == 'diff':
                feature_list.append(ts.max() - ts.min())
            else:
                print("Unrecognized feature passed in:", f)
                return None
    return feature_list

def extract_pbts_features_normalized (obj_id, df, feats = ['mean', 'stddev', 'med', 'max', 'min', 'kurt', 'skew', 'diff']):

    feature_list = [obj_id]
    for i, f in enumerate(feats):
        for name, obj_ts in df.groupby('passband'):
            ts = obj_ts.loc[:, 'flux'].copy()
            
            #normalize
            mean = ts.mean()
            std = ts.std()
            
            ts = (ts-mean)/std
            
            if f == 'mean':
                feature_list.append(ts.mean())
            elif f == 'stddev':
                feature_list.append(ts.std())
            elif f == 'med':
                feature_list.append(ts.median())
            elif f == 'max':
                feature_list.append(ts.max())
            elif f == 'min':
                feature_list.append(ts.min())
            elif f == 'kurt':
                feature_list.append(ts.kurtosis())
            elif f == 'skew':
                feature_list.append(ts.skew())
            elif f == 'diff':
                feature_list.append(ts.max() - ts.min())
            else:
                print("Unrecognized feature passed in:", f)
                return None
    return feature_list

def extract_features(ts_df, metadata, f_names = ['mean', 'stddev', 'med', 'max', 'min', 'kurt', 'skew', 'diff'], normalized = False):
    start = time.time()
    colnames = ['object_id']
    print('Extracting time series features for', metadata.shape[0], 'objects...', end = '')
    for f in f_names:
        for i in range(0, 6):
            colnames.append(f + str(i))

    feature_data = []
    obj_ids = metadata['object_id']
    groups = ts_df.groupby('object_id')
    for id in obj_ids:
        #print('Extracting features for:', id)
        obj_ts = groups.get_group(id)
        if normalized == False:
            feats = extract_pbts_features(id, obj_ts, feats = f_names)
        else:
            feats = extract_pbts_features_normalized(id, obj_ts, feats=f_names)
        feature_data.append(feats)

    ts_features = pd.DataFrame(feature_data, columns = colnames)
    end = time.time()
    print('done. Elapsed time was', end-start, 'seconds.' )
    
    return ts_features

#found a faster way with aggregate()
def extract_features2(ts_df, metadata, f_names = ['mean', 'stddev', 'med', 'max', 'min', 'kurt', 'skew', 'diff'], normalized = False):
    start = time.time()
    print('Extracting time series features for', metadata.shape[0], 'objects...', end = '')

    aggregate_dict = {
        'flux': ['mean', 'std', 'median', 'max', 'min'],
        'flux_err': ['mean', 'std', 'median', 'max', 'min'],
        'detected': ['mean']

    }

    colnames = [f+'_' + agg + str(i)  for i in range(0, 6) for f in aggregate_dict.keys() for agg in aggregate_dict[f]   ]

    feats0 = ts_df[ts_df['passband'] == 0].groupby(['object_id']).aggregate(aggregate_dict)
    feats0.columns = feats0.columns.droplevel(0)
    feats1 = ts_df[ts_df['passband'] == 1].groupby(['object_id']).aggregate(aggregate_dict)
    feats1.columns = feats1.columns.droplevel(0)
    feats2 = ts_df[ts_df['passband'] == 2].groupby(['object_id']).aggregate(aggregate_dict)
    feats2.columns = feats2.columns.droplevel(0)
    feats3 = ts_df[ts_df['passband'] == 3].groupby(['object_id']).aggregate(aggregate_dict)
    feats3.columns = feats3.columns.droplevel(0)
    feats4 = ts_df[ts_df['passband'] == 4].groupby(['object_id']).aggregate(aggregate_dict)
    feats4.columns = feats4.columns.droplevel(0)
    feats5 = ts_df[ts_df['passband'] == 5].groupby(['object_id']).aggregate(aggregate_dict)
    feats5.columns = feats5.columns.droplevel(0)

    #merge everything
    feats = feats0.merge(feats1, on='object_id')
    feats = feats.merge(feats2, on='object_id')
    feats = feats.merge(feats3, on='object_id')
    feats = feats.merge(feats4, on='object_id')
    feats = feats.merge(feats5, on='object_id')
    feats.columns = colnames
    feats = feats.reset_index()
    feats = feats.merge(metadata, on = 'object_id')
    feats = feats.fillna(0)
    labels = feats.loc[:, 'target']
    feats.drop('target', inplace = True, axis = 1)
    feats.drop('object_id', inplace = True, axis = 1)
    end = time.time()
    print('done. Elapsed time was', end-start, 'seconds.' )
    
    return feats, labels

def extract_features3(ts_df, metadata):
    start = time.time()
    #print('Extracting time series features for', metadata.shape[0], 'objects...', end = '')
    aggregate_dict = {
        'flux': ['mean', 'std', 'median', 'max', 'min'],
        'flux_err': ['mean', 'std', 'median', 'max', 'min'],
        'detected': ['mean']
    }

    colnames = [f+'_' + agg + str(i)  for i in range(0, 6) for f in aggregate_dict.keys() for agg in aggregate_dict[f]   ]

    feats0 = ts_df[ts_df['passband'] == 0].groupby(['object_id']).aggregate(aggregate_dict)
    feats0.columns = feats0.columns.droplevel(0)
    feats1 = ts_df[ts_df['passband'] == 1].groupby(['object_id']).aggregate(aggregate_dict)
    feats1.columns = feats1.columns.droplevel(0)
    feats2 = ts_df[ts_df['passband'] == 2].groupby(['object_id']).aggregate(aggregate_dict)
    feats2.columns = feats2.columns.droplevel(0)
    feats3 = ts_df[ts_df['passband'] == 3].groupby(['object_id']).aggregate(aggregate_dict)
    feats3.columns = feats3.columns.droplevel(0)
    feats4 = ts_df[ts_df['passband'] == 4].groupby(['object_id']).aggregate(aggregate_dict)
    feats4.columns = feats4.columns.droplevel(0)
    feats5 = ts_df[ts_df['passband'] == 5].groupby(['object_id']).aggregate(aggregate_dict)
    feats5.columns = feats5.columns.droplevel(0)

    #merge everything
    feats = feats0.merge(feats1, on='object_id')
    feats = feats.merge(feats2, on='object_id')
    feats = feats.merge(feats3, on='object_id')
    feats = feats.merge(feats4, on='object_id')
    feats = feats.merge(feats5, on='object_id')
    feats.columns = colnames
    feats = feats.reset_index()
    feats = feats.merge(metadata, on = 'object_id')
    feats = feats.fillna(0)
    feats.drop('object_id', inplace = True, axis = 1)
    end = time.time()
    #print('done. Elapsed time was', end-start, 'seconds.' )
    return feats

def addFeatsToMetadata(metadata, feats):
    df_ts_features = feats.iloc[:, 1:len(feats.columns)]
    df_metadata = metadata.iloc[:, 1:len(metadata.columns)]
    df_metadata = df_metadata.reset_index(drop = True)
    df_metadata = df_metadata[['ra', 'decl', 'gal_l', 'gal_b', 'ddf', 'hostgal_specz', 'hostgal_photoz', 'hostgal_photoz_err', 'distmod', 'mwebv']].copy()
    
    #augmented feature dataframe
    features = pd.concat([df_ts_features, df_metadata], axis = 1 )
    features = features.fillna(0)
    
    return features


def cv_curve(parameters, x_train, y_train, x_test, y_test):
    cv_avg_errors = []
    test_errors = []
    train_errors = []
    for p in parameters:
        print('running cv for p =', p)
        model = RandomForestClassifier(n_estimators = p, n_jobs = -1)
        cv_scores = cross_val_score(model, x_train, y_train, cv = StratifiedKFold(5))
        avg = sum(cv_scores)/5.0
        cv_avg_errors.append(1.0-avg)

    
    fig, ax = plt.subplots()
    plt.sca(ax)
    plt.plot(parameters, cv_avg_errors, '-gD')
    plt.legend(['avg cv error'])
    plt.xlabel('n_estimators')
    plt.ylabel('error rate')
    plt.ylim(0, 1)
    plt.show()
    plt.savefig('./figures/cv_curve.png')
    return cv_avg_errors
    

def cv_curve2(parameters, x_train, y_train, x_test, y_test):
    cv_avg_errors = []
    test_errors = []
    train_errors = []
    for p in parameters:
        print('running cv for min_number_samples =', p)
        model = RandomForestClassifier(n_estimators = 400, min_samples_split = p, n_jobs = -1)
        cv_scores = cross_val_score(model, x_train, y_train, cv = StratifiedKFold(5))
        avg = sum(cv_scores)/5.0
        cv_avg_errors.append(1.0-avg)
        
    fig, ax = plt.subplots()
    plt.sca(ax)
    plt.plot(parameters, cv_avg_errors, '-gD')
    plt.legend(['avg cv error'])
    plt.xlabel('min_sample_split')
    plt.ylabel('error rate')
    plt.ylim(0, 1)
    plt.show()
    plt.savefig('./figures/cv_curve2.png')
    return cv_avg_errors

def gaussian_naive_bayes_cv(x_train, y_train):
    
    gnb = GaussianNB()

    cv_scores = cross_val_score(gnb, x_train, y_train, cv = StratifiedKFold(5))

    print('----------------------------------------------------------------')
    print('|****              Gaussian Naive Bayes                     ****|')
    print('cv scores: ', *cv_scores.round(5), sep = ' ')
    print('cv average:', sum(cv_scores)/5.0)

    gnb.fit(x_train, y_train)
    
    return gnb, sum(cv_scores)/5.0

def logistic_regression_cv(x_train, y_train, l = 'l1', C = 10, max_iter = 1000, solver = 'liblinear', n_jobs = -1):
    
    log = LogisticRegression(penalty = l, C = C, solver = solver, max_iter = 1000, multi_class = 'ovr', n_jobs = n_jobs)

    cv_scores = cross_val_score(log, x_train, y_train, cv = StratifiedKFold(5))
    cv_avg = sum(cv_scores)/5.0
    print('----------------------------------------------------------------')
    print('|****LogisticRegression with', l, 'penalty', 'and C = ', C, '****|')
    print('cv scores: ', *cv_scores.round(5), sep = ' ')

    log.fit(x_train, y_train)
    
    return log, cv_avg

def decision_tree_cv(x_train, y_train):
    
    dt = DecisionTreeClassifier()

    cv_scores = cross_val_score(dt, x_train, y_train, cv = StratifiedKFold(5))

    print('----------------------------------------------------------------')
    print('|****                Decision    Tree                     ****|')
    print('cv scores: ', *cv_scores.round(5), sep = ' ')
    print('cv average:', sum(cv_scores)/5.0)

    dt.fit(x_train, y_train)
    
    return dt, sum(cv_scores)/5.0

def random_forest_cv(x_train, y_train, nest, min_split):
    
    rf = RandomForestClassifier(n_estimators = nest, min_samples_split = min_split)

    cv_scores = cross_val_score(rf, x_train, y_train, cv = StratifiedKFold(5))

    print('----------------------------------------------------------------')
    print('|****                Random    Forest                      ****|')
    print('cv scores: ', *cv_scores.round(5), sep = ' ')
    print('cv average:', sum(cv_scores)/5.0)

    rf.fit(x_train, y_train)
    
    return rf, sum(cv_scores)/5.0
def split_random_forest_cv(x_train, y_train, nest, min_split):
    
    ig_rf = RandomForestClassifier(n_estimators = nest, min_samples_split = min_split, n_jobs = -1)
    eg_rf = RandomForestClassifier(n_estimators = nest, min_samples_split = min_split, n_jobs = -1)
    skf = StratifiedKFold(n_splits = 5)
    
    total_scores = []
    cv_scores = []
    
    for train_idx, test_idx in skf.split(x_train, y_train):
        train_data, train_labels = x_train[train_idx], y_train[train_idx]
        test_data, test_labels = x_train[test_idx], y_train[test_idx]
        
        ig_x_train, ig_y_train, eg_x_train, eg_y_train = separate_inter_extragalactic(train_data, train_labels)
        ig_x_test, ig_y_test, eg_x_test, eg_y_test= separate_inter_extragalactic(test_data, test_labels)
        
        #get ig score
        ig_rf.fit(ig_x_train, ig_y_train)
        eg_rf.fit(eg_x_train, eg_y_train)
        
        total_score = accuracy_score(ig_rf.predict(ig_x_test), ig_y_test, normalize = False) + accuracy_score(eg_rf.predict(eg_x_test), eg_y_test, normalize = False)
        accuracy = total_score / test_labels.shape[0]
        total_scores.append(total_score)
        cv_scores.append(accuracy)
        
        
        

    print('----------------------------------------------------------------')
    print('|****            Random    Forest:', ig_y_train.shape[0],' objects        ****|')
    print('cv scores for ig random forest: ', *cv_scores.round(5), sep = ' ')
    print('raw scores for ig random forest:', *total_scores, sep = ' ')
    print('cv average for ig random forest:', sum(cv_scores)/5.0)

    ig_rf.fit(x_train, y_train)
    
    eg_cv_scores = cross_val_score(eg_rf, eg_x_train, eg_y_train, cv = StratifiedKFold(5), n_jobs = -1)
    eg_raw scores = eg_cv_scores * eg_y_train.shape[0]
    

    print('----------------------------------------------------------------')
    print('|****            Random    Forest:', eg_y_train.shape[0],' objects        ****|')
    print('cv scores for ig random forest: ', *eg_cv_scores.round(5), sep = ' ')
    print('raw scores for ig random forest:', *eg_raw_scores.round(0), sep = ' ')
    print('cv average for ig random forest:', sum(eg_cv_scores)/5.0)

    ig_x_train, ig_y_train, eg_x_train, eg_y_train = separate_inter_extragalactic(x_train, y_train)
    
    if_rf.fit(ig_x_train, ig_y_train)
    eg_rf.fit(eg_x_train, eg_y_train)
    
    
    return ig_rf, eg_rf, ig_cv_scores, eg_cv_scores

# def split_random_forest_cv(x_train, y_train, nest, min_split):
    
#     ig_x_train, ig_y_train, eg_x_train, eg_y_train = separate_inter_extragalactic(x_train, y_train)
#     ig_rf = RandomForestClassifier(n_estimators = nest, min_samples_split = min_split)
#     eg_rf = RandomForestClassifier(n_estimators = nest, min_samples_split = min_split)

#     ig_cv_scores = cross_val_score(ig_rf, ig_x_train, ig_y_train, cv = StratifiedKFold(5), n_jobs = -1)
    
#     ig_raw scores = ig_cv_scores * ig_y_train.shape[0]
    

#     print('----------------------------------------------------------------')
#     print('|****            Random    Forest:', ig_y_train.shape[0],' objects        ****|')
#     print('cv scores for ig random forest: ', *ig_cv_scores.round(5), sep = ' ')
#     print('raw scores for ig random forest:', *ig_raw_scores.round(0), sep = ' ')
#     print('cv average for ig random forest:', sum(ig_cv_scores)/5.0)

#     ig_rf.fit(x_train, y_train)
    
#     eg_cv_scores = cross_val_score(eg_rf, eg_x_train, eg_y_train, cv = StratifiedKFold(5), n_jobs = -1)
#     eg_raw scores = eg_cv_scores * eg_y_train.shape[0]
    

#     print('----------------------------------------------------------------')
#     print('|****            Random    Forest:', eg_y_train.shape[0],' objects        ****|')
#     print('cv scores for ig random forest: ', *eg_cv_scores.round(5), sep = ' ')
#     print('raw scores for ig random forest:', *eg_raw_scores.round(0), sep = ' ')
#     print('cv average for ig random forest:', sum(eg_cv_scores)/5.0)

#     eg_rf.fit(x_train, y_train)
    
    
#     return ig_rf, eg_rf, ig_cv_scores, eg_cv_scores

def adaboost_cv(x_train, y_train):
    
    ab = AdaBoostClassifier(DecisionTreeClassifier(max_depth = 2), n_estimators = 100)

    cv_scores = cross_val_score(ab, x_train, y_train, cv = StratifiedKFold(5))

    print('----------------------------------------------------------------')
    print('|****                Adaboost with DT base                 ****|')
    print('cv scores: ', *cv_scores.round(5), sep = ' ')
    print('cv average:', sum(cv_scores)/5.0)

    ab.fit(x_train, y_train)
    
    return ab, sum(cv_scores)/5.0

def create_logger():
    logger_ = logging.getLogger('main')
    logger_.setLevel(logging.DEBUG)
    fh = logging.FileHandler('simple_lightgbm.log')
    fh.setLevel(logging.DEBUG)
    ch = logging.StreamHandler()
    ch.setLevel(logging.DEBUG)
    formatter = logging.Formatter('[%(levelname)s]%(asctime)s:%(name)s:%(message)s')
    fh.setFormatter(formatter)
    ch.setFormatter(formatter)
    # add the handlers to the logger
    logger_.addHandler(fh)
    logger_.addHandler(ch)
    

def get_logger():
    return logging.getLogger('main')

def train_model(training_features, training_labels):
    
    
    rf = RandomForestClassifier(n_estimators = 500)
    cv_scores = cross_val_score(rf, training_features, training_labels, cv = StratifiedKFold(5))
    cv_avg = sum(cv_scores)/5.0
    
    print('----------------------------------------------------------------')
    print('|****               Random Forest Classifier               ****|')
    print('cv scores: ', *cv_scores.round(5), sep = ' ')

    rf.fit(training_features, training_labels)
    rf_score = rf.score(test_features, test_labels)
    print('Dtest score:', round(rf_score, 5))
    print('----------------------------------------------------------------\n\n')    
    
    return rf

def the_unique(x):
    return [x[i] for i in range(len(x)) if x[i] != x[i-1]]


def predict(ts, metadata, model):
    
    #extract features
    metadata = metadata.reset_index(drop = True)
    #display(metadata)
    features = extract_features3(ts, metadata)
    probs = model.predict_proba(features)
    probs_df = pd.DataFrame(probs, columns = ['class_' + str(c) for c in model.classes_])
    probs_df = probs_df.reset_index(drop = True)
    probs_df['class_99'] = (1.0 - probs).prod(axis = 1)
    probs_df.insert(0, 'object_id', metadata['object_id'])
    #display(probs_df)
    
    in_count = metadata.shape[0]
    out_count = probs_df.shape[0]
    first_in = metadata.iloc[0, 0]
    first_out = probs_df.iloc[0,0]
    if (first_in != first_out):
        print('something happened when reading in the ts starting with', ts.iloc[0, 0], '. first in =', first_in, 'first out =', first_out)
    if (in_count !=out_count):
        print('something happened when reading in the ts starting with', ts.iloc[0, 0])
    gc.collect()
    return probs_df
    
def predict_test_set_probs(model):
    
    start = time.time()
    remain_df = None
    chunks = 4000000
    
    test_set_md = pd.read_csv('../data/test_set_metadata.csv')
    test_set_md.sort_values(by = 'object_id' , inplace = True, axis = 0)
    test_set_md = test_set_md.reset_index(drop = True)
    
    count = 0
    rdf_size = -1
    for i_c, df in enumerate(pd.read_csv('../data/test_set.csv', chunksize=chunks, iterator=True)):
        # Check object_ids
        # I believe np.unique keeps the order of group_ids as they appear in the file
        # My belief is wrong (I should have read the doc !)
        # A big thank you to https://www.kaggle.com/filby89
        # Use .tolist() is almost 3 times faster than the_unique(df['object_id'].values)
        unique_ids = the_unique(df['object_id'].tolist())
        new_remain_df = df.loc[df['object_id'] == unique_ids[-1]].copy()



        if remain_df is None:
            df = df.loc[df['object_id'].isin(unique_ids[:-1])].copy()
        else:
            new_df = df.loc[df['object_id'].isin(unique_ids[:-1])].copy()
            df = pd.concat([remain_df, new_df], axis=0)
            if (df.shape[0] != (remain_df.shape[0] + new_df.shape[0])):
                print('!!!!!!!! size of df is', df.shape[0], '. should be', remain_df.shape[0] + new_df.shape[0])
        # Create remaining samples df
        remain_df = new_remain_df.copy()
        probs_df = predict(ts= df, metadata = test_set_md.loc[test_set_md['object_id'].isin(unique_ids[:-1])], model = model)
        #display(probs_df)
        count = count + probs_df.shape[0]
        if i_c == 0:
            probs_df.to_csv('../submissions/submission.csv', header=True, index=False, float_format='%.6f')
        else:
            probs_df.to_csv('../submissions/submission.csv', header=False, mode='a', index=False, float_format='%.6f')
        print('Written', count, 'rows out of 3492890')
        del probs_df
        gc.collect()

        if (i_c + 1) % 10 == 0:
            get_logger().info('%15d done in %5.1f' % (chunks * (i_c + 1), (time.time() - start) / 60))
            print('%15d done in %5.1f' % (chunks * (i_c + 1), (time.time() - start) / 60))

    # Compute last object in remain_df
#     display(unique_ids[-1])
#     display(test_set_md.loc[test_set_md['object_id'].isin(unique_ids[:-1])])
#     display(remain_df)
    probs_df = predict(ts= remain_df, metadata = test_set_md.loc[test_set_md['object_id'].isin([unique_ids[-1]])], model = model)
    count = count + probs_df.shape[0]
    probs_df.to_csv('../submissions/submission.csv', header=False, mode='a', index=False, float_format='%.6f')
    print('the count is', count)
    
def separate_inter_extragalactic(features, labels):
    
    df = features.copy()
    df.loc[:, 'target'] = labels
    
    ig_df = df.loc[df['hostgal_specz'] == 0].copy()

    eg_df = df.loc[df['hostgal_specz'] != 0].copy()
    
    eg_labels = eg_df['target']
    ig_labels = ig_df['target']
    eg_df.drop('target', inplace = True, axis = 1)
    ig_df.drop('target', inplace = True, axis = 1)
    
    return ig_df, ig_labels, eg_df, eg_labels
```
