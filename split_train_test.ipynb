```python
import pandas as pd
import matplotlib as plt
from sklearn.model_selection import train_test_split

#need to read csv files
data_all = pd.read_csv('~/Desktop/Classes/ee660/data/training_set.csv')
metadata_all = pd.read_csv('~/Desktop/Classes/ee660/data/training_set_metadata.csv')

#split into intergalactic and extragalactic
ig_metadata_all = metadata_all.loc[metadata_all['hostgal_specz']==0]
eg_metadata_all = metadata_all.loc[metadata_all['hostgal_specz'] != 0]

#use subset of training data as test set because test labels are not included in kaggle data
#separate labels from metadata

ig_labels_all = pd.DataFrame(ig_metadata_all.iloc[:, 11], columns = ['target'])
ig_metadata_all = ig_metadata_all.drop(['target'], axis = 1)

ig_train_metadata, ig_test_metadata, ig_train_labels, ig_test_labels = train_test_split(ig_metadata_all, ig_labels_all, test_size = 0.2, random_state = 888) 

ig_train_metadata['target'] = ig_train_labels['target']
ig_test_metadata['target'] = ig_test_labels['target']

ig_test_time_series = data_all.loc[data_all['object_id'].isin(ig_test_metadata.loc[:, 'object_id'])]
ig_train_time_series = data_all.loc[data_all['object_id'].isin(ig_train_metadata.loc[:, 'object_id'])]


ig_train_time_series.to_csv(path_or_buf = '~/Desktop/Classes/ee660/data/ig_training_set.csv', index = False)
ig_test_time_series.to_csv(path_or_buf = '~/Desktop/Classes/ee660/data/ig_test_set.csv', index = False)
ig_train_metadata.to_csv(path_or_buf = '~/Desktop/Classes/ee660/data/ig_training_set_metadata.csv', index = False)
ig_test_metadata.to_csv(path_or_buf = '~/Desktop/Classes/ee660/data/ig_test_set_metadata.csv', index = False)

#extragalactic
eg_labels_all = pd.DataFrame(eg_metadata_all.iloc[:, 11], columns = ['target'])
eg_metadata_all = eg_metadata_all.drop(['target'], axis = 1)

eg_train_metadata, eg_test_metadata, eg_train_labels, eg_test_labels = train_test_split(eg_metadata_all, eg_labels_all, test_size = 0.2, random_state = 888) 

eg_train_metadata['target'] = eg_train_labels['target']
eg_test_metadata['target'] = eg_test_labels['target']

eg_test_time_series = data_all.loc[data_all['object_id'].isin(eg_test_metadata.loc[:, 'object_id'])]
eg_train_time_series = data_all.loc[data_all['object_id'].isin(eg_train_metadata.loc[:, 'object_id'])]

eg_train_time_series.to_csv(path_or_buf = '~/Desktop/Classes/ee660/data/eg_training_set.csv', index = False)
eg_test_time_series.to_csv(path_or_buf = '~/Desktop/Classes/ee660/data/eg_test_set.csv', index = False)
eg_train_metadata.to_csv(path_or_buf = '~/Desktop/Classes/ee660/data/eg_training_set_metadata.csv', index = False)
eg_test_metadata.to_csv(path_or_buf = '~/Desktop/Classes/ee660/data/eg_test_set_metadata.csv', index = False)

#all
labels_all = pd.DataFrame(metadata_all.iloc[:, 11], columns = ['target'])
metadata_all = metadata_all.drop(['target'], axis = 1)

train_metadata, test_metadata, train_labels, test_labels = train_test_split(metadata_all, labels_all, test_size = 0.2, random_state = 888) 

train_metadata['target'] = train_labels['target']
test_metadata['target'] = test_labels['target']

test_time_series = data_all.loc[data_all['object_id'].isin(test_metadata.loc[:, 'object_id'])]
train_time_series = data_all.loc[data_all['object_id'].isin(train_metadata.loc[:, 'object_id'])]

train_time_series.to_csv(path_or_buf = '~/Desktop/Classes/ee660/data/all_training_set.csv', index = False)
test_time_series.to_csv(path_or_buf = '~/Desktop/Classes/ee660/data/all_test_set.csv', index = False)
train_metadata.to_csv(path_or_buf = '~/Desktop/Classes/ee660/data/all_training_set_metadata.csv', index = False)
test_metadata.to_csv(path_or_buf = '~/Desktop/Classes/ee660/data/all_test_set_metadata.csv', index = False)
```
